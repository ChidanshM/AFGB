### 1. What is the difference between enclosing a list comprehension in square brackets and parentheses?
- Square brackets (`[]`) - creates a new list object - the results of evaluating the expression for each item in the iterable.
- Parentheses (`()`) - creates a generator object 
---


### 2. What is the relationship between generators and iterators?
- Generators are a specific type of iterator created using generator functions or generator expressions.
- They are a kind of iterable that can be iterated over to produce values dynamically. 
- Generators use the `yield` statement to produce values one at a time, allowing for memory-efficient evaluation of results.
- All generators are iterators, but not all iterators are generators. 
---


### 3. What are the signs that a function is a generator function?
- Generator functions are defined using the `def` keyword, like regular functions, but they contain one or more `yield` statements.
---


### 4. What is the purpose of a yield statement?
- The `yield` statement is used in generator functions to yield a value to the caller and preserving the function's state. When the function is called, it returns a generator object that can be iterated over.
- The value yielded by the `yield` statement is returned to the caller, and the function's state is suspended. The function can later be resumed from the same state, allowing it to continue execution from where it left off.
- This would result in returning a value every time instead of computing all at once and returning an iterable object such as a list.
---


### 5. What is the relationship between map calls and list comprehensions? Make a comparison and contrast between the two.
- Both map calls and list comprehensions are used to transform data by applying a function to each item in an iterable.

- Map calls apply a function to each item in the iterable and return an iterator that yields the results. WHereas list comprehensions creates a new list containing the results of applied function to each item.

- List comprehensions - more readability and conciseness - especially for simple transformations. Map calls may be more memory-efficient when dealing with large datasets.
   
- Additional filtering and nesting capabilities, allowing for more complex transformations  is supported by list comprehension.
