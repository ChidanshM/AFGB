### 1. What does RGBA stand for?
RGBA stands for Red, Green, Blue, and Alpha. Colour model in which colours using a combination of intensity values of red, green, and blue, and an alpha channel for transparency.
___


### 2. From the `Pillow` module, how do you get the RGBA value of any images?
`getpixel()` method of the `Image` object to get the RGBA value of a specific pixel
```python
from PIL import Image
img = Image.open('example.jpg')
rgba_value = img.getpixel((x, y)) #pixel at coordinates (x, y)
print(rgba_value)
```
___


### 3. What is a box tuple, and how does it work?
A box tuple in Pillow is a tuple representing a rectangular region in an image. It consists of four integer values: (left, upper, right, lower), defining the respective coordinates edges. But the region includes the left and upper edges whereas excludes the right and lower edges.
___


### 4. Use your image and load in the notebook then, How can you find out the width and height of an `Image` object?
Use the `size` attribute of the `Image` object to find out its width and height.
```python
from PIL import Image
img = Image.open('example.jpg')
width, height = img.size
print(f"Width: {width}, Height: {height}")
```
___


### 5. What method would you call to get Image object for a 100×100 image, excluding the lower-left quarter of it?
You can use the `crop()` method to get an Image object for a specified region. For example:
```python
# Crop the image to exclude the lower-left quarter
new_img = img.crop((0, 0, width // 2, height // 2))
```
___


### 6. After making changes to an Image object, how could you save it as an image file?
`save()` method
```python
new_img.save('modified_example.jpg')
```

### 7. What module contains Pillow’s shape-drawing code?
`ImageDraw` module, to draw various shapes (lines, rectangles, circles, etc.) on an image.
___


### 8. Image objects do not have drawing methods. What kind of object does? How do you get this kind of object?
The `ImageDraw.Draw` object has drawing methods. You can obtain an `ImageDraw.Draw` object by calling the `ImageDraw.Draw()` function, passing in the Image object you want to draw on.
```python
from PIL import Image, ImageDraw
img = Image.open('example.jpg')
draw = ImageDraw.Draw(img) # Now you can use drawing methods on the draw object
```
