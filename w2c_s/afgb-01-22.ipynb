#### 1. What is the result of the code, and explain?
```py
X = 'iNeuron'
def func():
	print(X)
func()
```
**output:**
```
iNeuron
```
>**Explanation**: on calling `func()` it prints variable `X`, as no local variable are found it looks for global variable.

#### 2. What is the result of the code, and explain?
```py
X = 'iNeuron'
def func():
	X = 'NI!'
func()
print(X)
```
**output:**
```
iNeuron
```
>**Explanation**: on calling `func()` it initialise local variable `X`, in next line `print()` statement prints the global variable `X` initialised with `'iNeuron'`.

#### 4. What output does this code produce? Why?
```py
X = 'iNeuron'
def func():
	global X
	X='NI'
func()
print(X)
```
**output:**
```
NI
```
>**Explanation**: on calling `func()` it initialise global variable `X` with `'NI'` and in next line `print()` statement prints the global variable `X`.

#### 5. What about this code—what’s the output, and why?
```py
X = 'iNeuron'
def func():
	X = 'NI'
	def nested():
		print(X)
	nested()
func()
```
**output:**
```
NI
```
>**Explanation**: on calling `func()` it initialise local variable `X` with `'NI'` and then `nested()` is called which prints the variable `X` as no variable are found within the function now it looks for `X` in the enclosing scope.
---

#### 6. How about this code: what is its output in Python 3, and explain?
```py
def func():
	X = 'NI'
	def nested():
		nonlocal X
		X = 'Spam'
	nested()
	print(X)
func()
```
**output:**
```
Spam

```
>**Explanation**: on calling `func()` it initialise local variable `X` with `'NI'` and then `nested()` is called which the nonlocal `X` is defined referring to the `X` in the next enclosing scope,i.e., the `func()` block and with that this `X` is initialised with `'Spam'`. at the end of `func()` the local variable `X` is printed.
