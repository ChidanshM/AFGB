### 1. What is the relationship between def statements and lambda expressions ?
- Both `def` statements and `lambda` expressions are used to create functions in Python.
- `def` statements are used to define named functions with a block of code which can contain multiple statements and have a return value. Generally they are preferred for more complex function. 
- `lambda` expressions create anonymous functions (functions without a name) and are often used for conciseness operations, which are limited  to single expression. Generally they are used for simpler functions or as an arguments to higher-order functions like `map()`, `filter()`, and `sorted()`. 
---


### 2. What is the benefit of lambda?
- The main benefit of `lambda` expressions is to create anonymous functions with conciseness operations
- They are often used as a temporary simple function or as an arguments to higher-order functions.

### 3. Compare and contrast map, filter, and reduce.
- `map()`: function applied to all items in an input iterable and returns a new iterable containing the results.
- `filter()`: Filters elements from an input iterable based on a given function (predicate) and returns a new iterable containing only the elements for which the function returns `True`.
- `reduce()`: function applied of two arguments cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value.Take in note its a non-built-in function in Python 3 and must be imported from the `functools` module.
---


### 4. What are function annotations, and how are they used?
- An arbitrary metadata attached to function parameters and return values using the `->` syntax.
- Used to specify the expected types of function arguments and return values, but type checking not enforced.
- Purely optional and are mainly used for documentation purposes or by external tools.
- Example: `def add(x: int, y: int) -> int: return x + y`
---


### 5. What are recursive functions, and how are they used?
- Functions that call themselves either directly or indirectly in order to solve a problem.
- Often used to solve problems that can be broken down into smaller, similar sub-problems.
- They must have a base case to terminate the recursion, preventing infinite recursion.
- Example: Factorial function: `def factorial(n): return 1 if n == 0 else n * factorial(n - 1)`
---


### 6. What are some general design guidelines for coding functions?
- Keep functions concise and focused on a single task (single responsibility principle).
- Meaningful names - functions and parameters - to reflect their purpose.
- Document functions - docstrings - explain their purpose, parameters, and return values.
- Write testable functions by separating logic from I/O operations.
- Follow the DRY (Don't Repeat Yourself) principle and refactor repetitive code into reusable functions.
- Use function annotations for clarity, especially in larger codebases.
