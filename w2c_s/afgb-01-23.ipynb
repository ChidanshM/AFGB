### 1. What is the result of the code, and why?
```py
def func(a, b=6, c=8):
	print(a, b, c)
func(1, 2)
```
**output:**
```
1 2 8
```
> **Explanation**:  the function `func()` is called with arguments `1` and `2` assigned to `a` and `b`, thus for `c` the default value is used.

### 2. What is the result of this code, and why?
```py
def func(a, b, c=5):
	print(a, b, c)
func(1, c=3, b=2)
```
**output:**
```
1 2 3
```
> **Explanation**:  in the `func()` call values to arguments `c` and `b` are passed explicitly that's why the order of arguments isn't necessary, whereas `1` is passed and assigned to `a`

### 3. How about this code: what is its result, and why?
```py
def func(a, *pargs):
	print(a, pargs)
func(1, 2, 3)
```
**output:**
```
1 (2,3)
```
> **Explanation**: the `*pargs`is used to pass a non-keyworded variable length argument list to the function `func()` which then takes this list into a tuple. So during `func()` call `1` is assigned to `a` as it takes only a single value, then `2` and `3` assigned to `*pargs`
---


### 4. What does this code print, and why?
```py
def func(a, **kargs):
	print(a, kargs)
func(a=1, c=3, b=2)
```
**output:**
```
1 {'c': 3, 'b': 2}
```
> **Explanation**:  the `*pargs` parameter is used to pass keyworded variable length of arguments to the function, creating a dictionary. So, when `func(a=1, c=3, b=2)` is called, `a` is assigned `1`, and `kargs` is assigned `{'c': 3, 'b': 2}`.
---


### 5. What gets printed by this, and explain?
```py 
def func(a, b, c=8, d=5): 
	print(a, b, c, d)
func(1, *(5, 6))
```
**output:**
```
1 5 6 5
```
> **Explanation**: The `*` is an unpacking operator that unpacks the tuple `(5, 6)` into positional arguments. So, the call `func(1, *(5, 6))` is equivalent to `func(1, 5, 6)`, and the function prints `1 5 6 5.
---


### 6. what is the result of this, and explain?
```py
def func(a, b, c): a = 2; b[0] = 'x'; c['a'] = 'y'

l=1; m=[1]; n={'a':0}
func(l, m, n)
l, m, n
```
**output:**
```
(1, ['x'], {'a': 'y'})
```
> **Explanation**:  During `func()` call `l`, `m`, and `n` are passed as arguments. Inside `func()`, `a` is assigned the value `2`, the first element of list `b` is changed to `'x'`, and the value associated with key `'a'` in dictionary `c` is changed to `'y'`. Since lists and dictionaries are mutable, the changes made to `m` and `n` are reflected outside the function. Hence, after calling `func()`, `l` remains `1`, `m` becomes `['x']`, and `n` becomes `{'a': 'y'}`.
