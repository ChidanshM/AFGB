### 6. What is the difference between a `Run` object and a `Paragraph` object?
- A `Paragraph` object represents a paragraph of text in a Word document.
- A `Run` object is a part of a paragraph and represents a contiguous run of text with the same style such as inline content; text, pictures, and other items that are flowed between the block-item boundaries.
