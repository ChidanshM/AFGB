### 7. Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to its next() method: 2, 3, 5, 7, 11, ...

```python
def genPrimes():
	primes = [2, 3]
	yield 2
	yield 3
	n = 5
	while True:
		if all(n % p != 0 for p in primes):
			yield n
			primes.append(n)
		n += 2
```

```python
primes_generator = genPrimes()
print(next(primes_generator))  # Output: 2
print(next(primes_generator))  # Output: 3
print(next(primes_generator))  # Output: 5
# Continue calling next() to get the next prime numbers in the sequence
```
