- Logging messages can be redirected to different outputs, such as files.
- Logging messages can be categorised

### 9. What are the differences between the Step Over, Step In, and Step Out buttons in the debugger?
- **Step Over:** Executes the current line of code and stops at the next line. If the next line is a function call, it won't go into the function.
- **Step In:** Executes the current line of code and stops at the first line of any function that is called.
- **Step Out:** Continues execution until the current function returns.
___


### 10. After you click Continue, when will the debugger stop ?
The debugger will stop when it encounters a breakpoint, an exception is raised, or the program completes.
___


### 11. What is the concept of a breakpoint?
A breakpoint is a specified point in the code where the debugger will pause execution, allowing you to inspect variables, step through code, and analyze the program's state.
